**PROBLEM 1**

```python
# input layer
x1, x2, x3, x4 = 1, 0, 1, 0

# weights
w11, w12, w13 = 0.8, 0.3, -0.6
w21, w22, w23 = 0.8, 0.8, 0.3
w31, w32, w33 = 0.3, 0.8, 0.8
w41, w42, w43 = -0.6, 0.3, 0.8

# Thresholds
thd_1 = 0.5
thd_2 = 0.3

# Hidden layers formulas
y1 = (x1 * w11) + (x2 * w21) + (x3 * w31) + (x4 * w41)
y2 = (x1 * w12) + (x2 * w22) + (x3 * w32) + (x4 * w42)
y3 = (x1 * w13) + (x2 * w23) + (x3 * w33) + (x4 * w43)

# Function to determine if neuron fires
def check_firing(y, threshold):
    return y >= threshold

# Results for threshold >= 0.5
firing_0_5 = {
    'y1': check_firing(y1, thd_1),
    'y2': check_firing(y2, thd_1),
    'y3': check_firing(y3, thd_1)
}

# Results for threshold >= 0.3
firing_0_3 = {
    'y1': check_firing(y1, thd_2),
    'y2': check_firing(y2, thd_2),
    'y3': check_firing(y3, thd_2)
}

# result display
print("Results for threshold >= 0.5:")
print(f"y1 = {y1:.1f}, fires: {firing_0_5['y1']}")
print(f"y2 = {y2:.1f}, fires: {firing_0_5['y2']}")
print(f"y3 = {y3:.1f}, fires: {firing_0_5['y3']}")

print("\nResults for threshold >= 0.3:")
print(f"y1 = {y1:.1f}, fires: {firing_0_3['y1']}")
print(f"y2 = {y2:.1f}, fires: {firing_0_3['y2']}")
print(f"y3 = {y3:.1f}, fires: {firing_0_3['y3']}")
```

**PROBLEM 4**

```python
A = "01100110111000101010"
B = "11011000110100011011"
start_loci = 5
end_loci = 8

# Function to perform crossover by swapping bits between A and B from loci 6-9 (both inclusive)
def crossover(A, B, start, end):
    # For better comprehension, split the chromosomes into three parts: before, crossover section, and after
    A_part1 = A[:start]
    A_crossover = A[start:end+1]
    A_part3 = A[end+1:]
    
    B_part1 = B[:start]
    B_crossover = B[start:end+1]
    B_part3 = B[end+1:]
    
    # Swapping the crossover sections
    new_A = A_part1 + B_crossover + A_part3
    new_B = B_part1 + A_crossover + B_part3
    
    return new_A, new_B

def display_crossover(A, new_A, B, new_B, start, end):
    def format_chromosome(chromosome, start, end):
        return chromosome[:start] + " " + chromosome[start:end+1] + " " + chromosome[end+1:]
        
    formatted_A = format_chromosome(A, start, end)
    formatted_new_A = format_chromosome(new_A, start, end)
    formatted_B = format_chromosome(B, start, end)
    formatted_new_B = format_chromosome(new_B, start, end)
    
    # Special display to see differences in  visual comparison 
    print("A:     ", formatted_A)
    print("New A: ", formatted_new_A)
    print("B:     ", formatted_B)
    print("New B: ", formatted_new_B)

new_A, new_B = crossover(A, B, start_loci, end_loci)

display_crossover(A, new_A, B, new_B, start_loci, end_loci)
```

**PROBLEM 5**

```python
A = "01100110111000101010"
B = "11011000110100011011"
C = "00011010111011110000"


# HAMMING DISTANCE function
def HD(x, y):
    assert len(x) == len(y)
    return sum(el1 != el2 for el1, el2 in zip(x, y))

dist_A_B = HD(A, B)
dist_A_C = HD(A, C)
dist_B_C = HD(B, C)

# result display
print(f"Hamming distance between A and B: {dist_A_B}")
print(f"Hamming distance between A and C: {dist_A_C}")
print(f"Hamming distance between B and C: {dist_B_C}")

```
